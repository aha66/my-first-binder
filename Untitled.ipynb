```python
import math
import cmath
import matplotlib.pyplot
import matplotlib.patches
import matplotlib.animation
import IPython.display
```

```python
x = (-0.2-0.8j, 0.4+0.6j)
```

```python
def line_coords(n,d):
    return [(0.,0.)] + [(k/(2*n), (k%4-2)*d) for k in range(1,2*n,2)] + [(1.,0.)]
```

```python
spring=line_coords(10,0.04)
```

```python
def coord(x, phi):
    fac = cmath.exp(complex(0,phi))
    return tuple((c*fac).real for c in x)
```

```python
def foo(x):
    u = sum(2*c.real*c.imag for c in x)
    v = sum(c.real*c.real-c.imag*c.imag for c in x)
    t = math.atan2(-u, v)/2
    p = [coord(x, t+w*math.pi/2) for w in range(4)]
    print(sum(c0*c1 for c0,c1 in zip(p[0],p[1])))
    a = math.hypot(*p[0])
    b = math.hypot(*p[1])
    phi = math.atan2(*p[0][::-1])
    return phi,a,b
```

```python
def animate(frame, nframes, point, x):
    point.center = coord(x, frame/nframes*2*math.pi)
    t = matplotlib.transforms.Affine2D().scale(math.hypot(*point.center),1).rotate(math.atan2(*point.center[::-1]))
    line.set_data(zip(*t.transform(spring)))
    return point,line
```

```python
ax = matplotlib.pyplot.gca()
line, = ax.plot(*list(zip(*spring)))
point = matplotlib.patches.Circle((0,0), 0.05)
ax.add_patch(point)
phi,a,b=foo(x)
ell = matplotlib.patches.Ellipse((0,0), 2*a, 2*b, math.degrees(phi), fill=False)
ax.add_patch(ell)
ax.set_aspect('equal')
ax.set_xbound(-1,1)
ax.set_ybound(-1,1)
```

```python
duration = 2000
interval = 30
n = duration//interval
anim=matplotlib.animation.FuncAnimation(point.figure, animate, n, fargs=(n,point,x), interval=30)
```
